# Performing Hyperparameter Tuning Of The Random Forest Model --

```python
#importing basic library
import pandas as pd
import numpy as np
```

```python
#importing and viewing data
df=pd.read_csv('C:\\Users\\AKASH pc\\Downloads\\archive (1)\\data.csv')
df.head()
```

```python
#looking at the shape of the data
df.shape
```

```python
#dropping unnessory column
df.drop('Unnamed: 32',axis=1,inplace=True)
```

```python
df.head()
```

```python
#dividing dependent and target features
X=df.drop('diagnosis',axis=1)
y=df.diagnosis
```

```python
#Importing train-test split
from sklearn.model_selection import train_test_split
```

```python
#performing train-test split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=.25,random_state=23)
```

```python
#using Randomforeset classifier as the ml technique
from sklearn.ensemble import RandomForestClassifier
```

```python
#fitting data into the model and making prediction 
rf=RandomForestClassifier().fit(X_train,y_train)
y_pred=rf.predict(X_test)
```

```python
#importing accuracy score to look how our model performed
from sklearn.metrics import accuracy_score
```

```python
#accuracy score of the model is
accuracy_score(y_test,y_pred)
```

That's a good accuracy. Let's see if we could increase the accuracy by hyperparamater tuning

# Doing Hyperparamter Tuning To Perform The Accuracy Of The Model

### 1.RandomisedSearchCV

```python
#importing randomisedSearchCV and making the parameter grid of the hyperparamter we want in our model
from sklearn.model_selection import RandomizedSearchCV
n_estimators = [int(x) for x in np.linspace(start = 200, stop = 2000, num = 6)]
max_features = ['auto', 'sqrt','log2']
max_depth = [int(x) for x in np.linspace(10, 1000,6)]
min_samples_split = [2,6,10]
min_samples_leaf = [1,3,6,8]
randomisedcv_param_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf,
              'criterion':['entropy','gini']}
```

```python

randomisedcv_param_grid
```

```python
#Fitting our data into the RandomisedSearchCv to look at the best hyperparameters for our model.It will take a lot time
res_cv=RandomizedSearchCV(rf,param_distributions=randomisedcv_param_grid,n_iter=100,verbose=2,
    n_jobs=1,cv=3).fit(X_train,y_train)
```

```python
#we can see that out accuracy score has been increased to some extinct and that's good
res_cv.best_score_
```

```python
#best paramters for our model
res_cv.best_params_
```

### 2. Grid Search

```python
from sklearn.model_selection import GridSearchCV
```

```python
#making gridsearch param grid from the result we obtain from RandomisedSearchCv
gridcv_param_grid={'n_estimators':[100,200,300],
 'min_samples_split': [1,2,3],
 'min_samples_leaf': [1,2,3],
 'max_features': ['auto','sqrt','log2'],
 'max_depth': [800,900,1000,1100,1200],
 'criterion': ['entropy','gini']}
```

```python
#looking at our param grid for GridSearchCV
gridcv_param_grid
```

```python
#fitting our model into the GridSearch to look at the best hyperparameter for our model.It will also take some time
res_cv2=GridSearchCV(rf,gridcv_param_grid,n_jobs=1,cv=3,verbose=2).fit(X_train,y_train)
```

```python
#looking at the accuracy score of our model now
res_cv2.best_score_
```

Yes, we have successfully increased the accuracy of our model to 97.2 from 95.8

```python
#the best hyperparamter for our model are-:
res_cv2.best_estimator_
```
